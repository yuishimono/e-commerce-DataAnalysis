```python
#ライブラリの読み込み
import pandas as pd
from sklearn.cluster import KMeans#sklearn.clusterという機能群の中にあるKMeans（クラスタリングのアルゴリズムの一つ）を使う
from sklearn.preprocessing import MinMaxScaler#sklearn.preprocessingという機能群の中にあるMinMaxScaler(正規化を行う)を使う
import pickle#Pythonで使用している情報をそのまま保存するためのライブラリ
```

```python
#初期定義
data_dir = '../data'
output_dir = f'{data_dir}/output'
input_file = f'{output_dir}/order_processed.csv'#加工済みデータのパス
output_file = f'{output_dir}/kmeans_result.csv'#クラスタリングの結果データのパス
model_file = f'{output_dir}/kmeans_model.pkl'#構築したモデルファイルのパス k-meansモデルをpickle形式で保存
```

```python
#データの読み込み
df = pd.read_csv(input_file)
display(df.head())
```

```python
#学習用にデータを加工
use_cols = ['顧客ID','売上','カテゴリ']
tg_df = df[use_cols]

display(tg_df)
```

```python
category = tg_df['カテゴリ'].unique()#カテゴリの一覧を取得

print(category)
```

```python
#変数の横持化
tg_df_yoko = []
for tg_category in category:
    tmp_df = tg_df.loc[tg_df['カテゴリ']==tg_category].copy()
    #tg_dfというデータフレームから該当する行を抽出して、新しいデータフレームtmp_dfを作成 ==は等しいことを意味する比較演算子　copy()は選択された行を新しいデータフレームとしてコピーするメソッド
    tmp_df[f'購買回数_{tg_category}'] = 1#tmp_dfに、f'購買回数_{tg_category}'という列を追加しています。この列は、該当する行がある場合は常に1となります
    tmp_df.rename(columns={'売上':f'売上_{tg_category}'},inplace = True)#tmp_dfの'売上'列の列名をf'売上_{tg_category}' に変更
    tg_df_yoko.append(tmp_df)#変換したtmp_dfを、リストtg_df_yokoに追加
tg_df_yoko = pd.concat(tg_df_yoko, ignore_index=False)#pd.concat()を使って、リストtg_df_yoko内のデータフレームを縦方向に連結し、最終的なデータフレームtg_df_yokoを作成

display(tg_df_yoko)
```

```python
#顧客単位での集計
X_df = tg_df_yoko.groupby('顧客ID').sum()

display(X_df)
```

```python
print(X_df.isnull().sum())
```

```python
#変数を整える
scaler = MinMaxScaler()#正規化を行うMinMaxScaler()を定義
X_df_scaled = scaler.fit_transform(X_df)#fit_transform()の引数で指定したX_dfのスケーリング処理を行う

print(X_df_scaled)
```

```python
model = KMeans(n_clusters=4, n_init=10, random_state=0).fit(X_df_scaled)
```

```python
pred = model.predict(X_df_scaled)
print(pred)#グループ番号を出力
```

```python
#モデルの評価
X_df['group'] = pred#クラスタリングによって作成したグループ番号を元々あったデータとくっつける
display(X_df.head())
```

```python
print(X_df['group'].value_counts())#各グループの件数を確認する
```

```python
#モデルの解釈をする
group_df = X_df.groupby('group').mean()#グループ単位の平均値を集計

display(group_df)
```

```python
group_rank_df = group_df.rank(ascending= False)#ランキングの作成、ascensing= Falseで売上の高いものがランクも高い
display(group_rank_df)
```

```python
#データの出力
X_df.to_csv(output_file)
```

```python
pickle.dump(model, open(model_file, 'wb'))
```
